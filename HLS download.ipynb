This code downloads HLS data. The example shows how to download the whole year of 2020 data (L30) for the geographic area (17TLF) that covers old woman creek wetland in Huron, OH.

```python
from bs4 import BeautifulSoup
import requests
r  = requests.get("https://hls.gsfc.nasa.gov/data/v1.4/L30/2020/17/T/L/F/")
data = r.text
soup = BeautifulSoup(data)
```

```python
for link in soup.find_all('a')[6:]:
    print(link.get('href')) 
    url = "https://hls.gsfc.nasa.gov/data/v1.4/L30/2020/17/T/L/F/"+ link.get('href')
    print(url)
    myfile = requests.get(url)
    store_path = "D:/HLS/2020/L30/"+link.get('href')
    print(store_path)
    open(store_path, 'wb').write(myfile.content)
```
